```python
#Genre wise reviews
data['Genre'].value_counts().plot(kind='bar',figsize=(5,5))
plt.title("User Rating V/s Genre")
plt.xlabel("Genre")
plt.ylabel("User Rating")
```

```python
#Year wise Review
data.groupby('Year')[[ 'Reviews']].mean().plot(kind='barh', figsize=(10,5))
```

```python
#Top Authors by user-rating
data1 = data['User Rating']
rating = max(data1)
print(rating)
Authors = data.Author[data['User Rating'] == rating]
print(Authors)
```
