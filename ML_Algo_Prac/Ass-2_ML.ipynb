```python
#Drop address, date, postcode, YearBuilt, lattitude, longtitude columns 
data.drop(['Address','Date','Postcode','YearBuilt','Lattitude','Longtitude'],axis = 'columns' , inplace = True)
data
```

```python
#Find the count of null value in each column 
data.isnull().sum()
```

```python
data
```

```python
#Fill the null value of property count, distance, Bedroom2, Bathroom, Car with 0
data['Propertycount'] = data.Propertycount.replace(np.NAN,0.0)
data['Distance'] = data.Distance.replace(np.NAN,0.0)
data['Bedroom2'] = data.Bedroom2.replace(np.NAN,0.0)
data['Bathroom'] = data.Bathroom.replace(np.NAN,0.0)
data['Car'] = data.Car.replace(np.NAN,0.0)
data
```

```python
#Q-7
#Fill Null value of land size and bidding area columns with Mean
land = data['Landsize'].loc[data['Landsize'] != np.NAN]
mean = land.astype(float).mean()
print(mean)
data['Landsize'] = data['Landsize'].replace(np.NAN,mean).astype(float)
data
```

```python
area = data['BuildingArea'].loc[data['BuildingArea'] != np.NAN]
mean = area.astype(float).mean()
print(mean)
data['BuildingArea'] = data['BuildingArea'].replace(np.NAN,mean).astype(float)
data.head()
```

```python
#Find the unique value in method column 
data.Method.unique()
```

```python
data1 = data.dropna(axis = 0)
```

```python
data1
```

```python
data1.isnull().sum()
```

```python
#Create the target data and feature data where target data is price
```

```python
X=data1.drop(['Price'],axis=1)
Y=data1.Price
```

```python
#Using Label Encoder for converting catergraical data into measure
```

```python
from sklearn.preprocessing import LabelEncoder
X=X.apply(LabelEncoder().fit_transform)
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(3)
X=poly.fit_transform(X)
```

```python
#Spliting Data
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=10)
```

```python
print("Shape:- ",X_train.shape)
print("Shape:- ",Y_train.shape)
print("Shape:- ",X_test.shape)
print("Shape:- ",Y_test.shape)
```

```python
#Model Fitting
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
lr=LinearRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
#Model Evalution
#Score
```

```python
lr.score(X_test,Y_test)
```

```python
Y_pred = lr.predict(X_test)
```

```python
Y_pred
```

```python
lr.score(X_train,Y_train)
```

```python
print('MSE: ',mean_squared_error(Y_test,Y_pred))
print('MAE: ',mean_absolute_error(Y_test,Y_pred))
print('R2_score: ',r2_score(Y_test,Y_pred))
```
